```python
import pandas as pd
d_comments = pd.read_csv("/media/tycho/Seagate Expansion Drive/subsets/subset_comments.csv",
                         sep="\t", header=0, nrows = 500000 )
```

```python
d_comments['label'] = d_comments['dataset'].apply(lambda x: 1 if x=="r" else 0)
x=d_comments[['comment',"dataset"]]
r= x[x["dataset"]=="r"];print(len(r))
l= x[x["dataset"]=="l"];print(len(l))
```

```python
from gensim.models import Word2Vec
import nltk

rr=[nltk.word_tokenize(str(w).lower()) for w in r["comment"] ] 
ll=[nltk.word_tokenize(str(w).lower()) for w in l["comment"] ]

word2vec_l = Word2Vec(ll, min_count=5)
word2vec_r = Word2Vec(rr, min_count=5)

word2vec_l.wv.most_similar("equal")
word2vec_r.wv.most_similar("equal")

positive = ["trump"]; negative=["obama"]
x_l = word2vec_l.wv.most_similar(positive ,negative,topn=20)
x_r = word2vec_r.wv.most_similar(positive ,negative,topn=20)
pd.concat([pd.Series(x_l),pd.Series(x_r)],keys = ["left","right"],axis=1)
```
