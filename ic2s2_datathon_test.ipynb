## ML models

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

d_comments2 = d_comments[["comment","label"]].dropna()
X_train,X_test,y_train,y_test = train_test_split(d_comments2["comment"],
                                                 d_comments2["label"],
                                                 random_state=0)


vect = CountVectorizer(min_df=20, max_df = .2,  
                        stop_words ="english", token_pattern='(?u)\\b\\w\\w\\w+\\b').fit(X_train)
```

```python
len(vect.get_feature_names())
```

```python
X_train_vectorized = vect.transform(X_train)

model = LogisticRegression()
model.fit(X_train_vectorized,y_train)
```

```python
from sklearn.metrics import roc_auc_score

predictions = model.predict(vect.transform(X_test))

roc_auc_score(y_test,predictions)
```

```python
feature_names = np.array(vect.get_feature_names())

sorted_coef_index = model.coef_[0].argsort()
print(feature_names[sorted_coef_index[:-100:-1]])
feature_names[sorted_coef_index[:100]]
```
