```python
import gensim
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vect =  CountVectorizer(min_df=20, max_df = .2,  
                        stop_words ="english", token_pattern='(?u)\\b\\w\\w\\w+\\b')
```

```python
r_tokens = [nltk.word_tokenize(msg) for msg in rr if type(msg)==str]
```

```python
X_r = vect.fit_transform([str(w) for w in rr])
X_l = vect.fit_transform([str(w) for w in ll])
```

```python
corpus_r = gensim.matutils.Scipy2Corpus(X_r)#,documents_columns = False)
id_map = dict((v,k) for k,v in vect.vocabulary_.items())
```

```python
corpus_l = gensim.matutils.Scipy2Corpus(X_l)
id_map = dict((v,k) for k,v in vect.vocabulary_.items())
```

```python
id_map
```

```python
ldamodel_r = gensim.models.ldamodel.LdaModel(corpus_r, num_topics = 5,
                                           id2word=id_map,
                                           update_every=5,
                                           chunksize=10000,
                                           passes=5)
```

```python
ldamodel_r.print_topics(num_topics=10,num_words=15)
```

```python
ldamodel_l = gensim.models.ldamodel.LdaModel(corpus_l, num_topics = 5,
                                           id2word=id_map,
                                           update_every=5,
                                           chunksize=10000,
                                           passes=10)
```

```python
(ldamodel_l.print_topics(num_topics=10,num_words=15))
```
